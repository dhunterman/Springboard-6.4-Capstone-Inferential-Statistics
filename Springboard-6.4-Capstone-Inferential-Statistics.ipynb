```python
import pandas as pd
import numpy as np
import datetime as dt

dfyear = pd.read_csv("data/fs_year_csv.csv")
dfyearmonth = pd.read_csv("data/fs_year_month_csv.csv")

print(dfyearmonth)

```

```python
dfyearm_i = dfyearmonth[dfyearmonth.Type=='Industrial']
dfyearm_i = dfyearm_i.rename(columns={'Year-Month':'YearMonth'})

dfyearm_i.YearMonth = pd.to_datetime(dfyearm_i.YearMonth)
dfyearm_i.YearMonth = dfyearm_i.YearMonth.dt.strftime('%m')
dfyearm_i2 = dfyearm_i.tail(60)
dfyearm_i2
```

```python
dfyearm_i3 = dfyearm_i2.copy()
dfyearm_i3 = dfyearm_i3[dfyearm_i3.columns[2:]].replace('[\$,]', '', regex=True).astype(float)
dfyearm_i4 = dfyearm_i2.copy()
dfyearm_i4['Americas']=dfyearm_i3['Americas']
dfyearm_i4['APAC']=dfyearm_i3['APAC']
dfyearm_i4['EMEA']=dfyearm_i3['EMEA']
dfyearm_i4['Total']=dfyearm_i3['Total']
#dfyearm_i3.head()

#dfyearm_i4 = dfyearm_i4.groupby(['YearMonth']).agg({'Americas': ['mean','min','max','std'],'Am Orders': ['mean','min','max','std'],'APAC':['mean','min','max','std'],'APAC Orders':['mean','min','max','std'],'EMEA':['mean','min','max','std'],'EMEA Orders':['mean','min','max','std'],'Total':['mean','min','max','std'],'Total Orders':['mean','min','max','std']}).reset_index()
dfyearm_am = dfyearm_i4.groupby(['YearMonth']).agg({'Americas': ['mean','min','max','std'],'Am Orders': ['mean','min','max','std']}).reset_index()
dfyearm_APAC = dfyearm_i4.groupby(['YearMonth']).agg({'APAC': ['mean','min','max','std'],'APAC Orders': ['mean','min','max','std']}).reset_index()
dfyearm_EMEA = dfyearm_i4.groupby(['YearMonth']).agg({'EMEA': ['mean','min','max','std'],'EMEA Orders': ['mean','min','max','std']}).reset_index()
dfyearm_total = dfyearm_i4.groupby(['YearMonth']).agg({'Total': ['mean','min','max','std'],'Total Orders': ['mean','min','max','std']}).reset_index()
```
